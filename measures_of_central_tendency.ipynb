# Who scores the most on average for math, reading and writing based on
 # Gender
 # Test preparation course

```python
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
dt=data.groupby('gender')['math score'].mean().plot.bar()
plt.title("Math score based on gender",fontsize=15)
plt.ylabel('Average math score', fontsize=10)
plt.xlabel('Gender', fontsize=15)

plt.subplot(1,3,2)
dt=data.groupby('gender')['reading score'].mean().plot.bar()
plt.title("Reading score based on gender",fontsize=15)
plt.ylabel('Average reading score', fontsize=10)
plt.xlabel('Gender', fontsize=15)

plt.subplot(1,3,3)
dt=data.groupby('gender')['writing score'].mean().plot.bar()
plt.title("Writing score based on gender",fontsize=15)
plt.ylabel('Average writing score', fontsize=10)
plt.xlabel('Gender', fontsize=15)
```

```python
#Based on gender male scores most in maths, females scores the most in reading and female scores most in writing
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
da=data.groupby('test preparation course')['math score'].mean().plot.bar()
plt.title("Math score based on test preparation course", fontsize=10)
plt.ylabel('Average math score', fontsize=12)
plt.xlabel('Test preparation course', fontsize=15)

plt.subplot(1,3,2)
da=data.groupby('test preparation course')['reading score'].mean().plot.bar()
plt.title("Reading score based on test preparation course", fontsize=10)
plt.ylabel('Average reading score', fontsize=12)
plt.xlabel('Test preparation course', fontsize=15)

plt.subplot(1,3,3)
da=data.groupby('test preparation course')['writing score'].mean().plot.bar()
plt.title("Writing score based on test preparation course", fontsize=10)
plt.ylabel('Average writing score', fontsize=12)
plt.xlabel('Test preparation course', fontsize=15)
```

```python
#Students who completed test preperation course score the most in maths, reading and writing
```
